## 1 - Introdução

### 1.1 - Objetivos do Projeto

1 - Objetivo principal do projeto:
* Desenvolver um sistema de engenharia de dados que permita coletar, processar e analisar os dados da Olist Brasil para obter insights valiosos e tomar decisões estratégicas.

2 - Objetivos específicos:

a) Análise de vendas:
* Analisar as vendas totais da Olist Brasil, identificando os produtos mais populares, tendências sazonais e períodos de maior demanda.
* Identificar as categorias de produtos que geram mais receita e margens de lucro mais altas.
* Identificar oportunidades para impulsionar as vendas, como promoções sazonais ou campanhas de marketing direcionadas a produtos específicos.

b) Análise de clientes:
* Compreender o perfil dos clientes da Olist Brasil, como localização geográfica, demografia e comportamento de compra.
* Segmentar os clientes com base em suas características e preferências, permitindo estratégias de marketing personalizadas e melhor atendimento às suas necessidades.
* Identificar clientes leais e criar programas de fidelidade para incentivar compras repetidas.

c) Análise de avaliações e feedback dos clientes:
* Analisar as avaliações e o feedback dos clientes para identificar padrões de satisfação, áreas de melhoria do produto ou serviço e oportunidades para aumentar a qualidade e a experiência do cliente.
* Identificar problemas recorrentes e implementar ações corretivas para melhorar a satisfação geral do cliente.
* Utilizar os insights dos clientes para aprimorar o desenvolvimento de novos produtos e serviços.

3 - Metas do projeto:

* Implementar um pipeline de ETL (Extração, Transformação e Carga) eficiente para coletar dados de diferentes fontes, como sistemas de vendas, mídias sociais e logs de atendimento ao cliente.
* Desenvolver um modelo de dados robusto e escalável para armazenar e organizar os dados da Olist Brasil.
* Criar visualizações de dados interativas e dashboards que permitam aos usuários explorar e analisar as informações de maneira intuitiva.
* Garantir a qualidade dos dados por meio de verificações de integridade, limpeza e enriquecimento dos dados.

4 - Resultados esperados:

* Obter insights acionáveis para melhorar a eficiência operacional, otimizar as estratégias de vendas e marketing e aprimorar a experiência do cliente.
* Identificar oportunidades de crescimento de negócios e tomar decisões baseadas em dados para impulsionar a lucratividade.
* Facilitar a tomada de decisões informadas em diferentes áreas da empresa, incluindo vendas, marketing, logística e desenvolvimento de produtos.

### 1.3 - Visão geral da Olist Brasil e seus desafios de dados

A Olist é uma empresa brasileira de tecnologia que atua como uma plataforma de comércio eletrônico, conectando pequenas e médias empresas a grandes marketplaces. A empresa foi fundada em 2015 e desde então tem desempenhado um papel importante no cenário do comércio eletrônico no Brasil.

A plataforma da Olist permite que as empresas parceiras listem e vendam seus produtos em diversos marketplaces populares, como Mercado Livre, Americanas, Submarino, entre outros. Dessa forma, as empresas podem ampliar sua visibilidade, alcançar um público maior e aumentar suas vendas.

Em relação aos desafios de dados, a Olist enfrenta alguns obstáculos comuns a empresas desse ramo, tais como:

Gerenciamento e integração de dados: 
* A Olist lida com uma grande quantidade de dados provenientes de várias fontes, incluindo informações dos vendedores, produtos, transações, avaliações dos clientes, entre outros. Um desafio é garantir a integridade, qualidade e consistência desses dados, bem como sua integração para fornecer uma visão abrangente das operações.

Análise de dados: 
* A empresa precisa extrair informações valiosas a partir dos dados coletados para tomar decisões estratégicas. Isso envolve a aplicação de técnicas de análise de dados, como mineração de dados, análise preditiva e geração de insights acionáveis. A Olist pode enfrentar desafios relacionados à capacidade de processar e analisar grandes volumes de dados em tempo hábil.

Personalização e recomendação: 
* A Olist busca oferecer uma experiência personalizada aos usuários, fornecendo recomendações relevantes de produtos com base em seus interesses e histórico de compras. Isso requer a implementação de algoritmos de recomendação e sistemas de filtragem colaborativa, o que pode ser desafiador devido à diversidade de produtos e à necessidade de processar um grande número de transações em tempo real.

Segurança e privacidade: 
* Como a Olist lida com informações confidenciais de seus clientes e parceiros, é essencial garantir a segurança e a privacidade desses dados. Isso envolve a implementação de medidas de segurança robustas, como criptografia de dados, autenticação de usuários e monitoramento de atividades suspeitas. Além disso, a empresa precisa estar em conformidade com regulamentações de proteção de dados, como a Lei Geral de Proteção de Dados (LGPD) no Brasil.

Escalabilidade e infraestrutura: 
* À medida que a Olist cresce e atrai mais parceiros e clientes, sua infraestrutura de dados precisa ser escalável para lidar com o aumento do volume e da complexidade dos dados. Isso requer investimentos em infraestrutura de TI, como servidores, armazenamento de dados e sistemas de processamento, além de garantir a disponibilidade e a confiabilidade desses recursos.

Qualidade dos dados: 
* A qualidade dos dados é fundamental para obter insights precisos e confiáveis. A Olist enfrenta o desafio de lidar com dados desestruturados e incompletos, bem como com a necessidade de garantir a consistência e a integridade dos dados ao longo do tempo. A implementação de processos de limpeza, padronização e validação de dados é essencial para melhorar a qualidade dos dados e evitar decisões errôneas baseadas em informações imprecisas.

Visualização e comunicação de dados: 
* Transformar dados complexos em informações compreensíveis e acionáveis é um desafio enfrentado pela Olist. A empresa precisa desenvolver dashboards e relatórios intuitivos e interativos para permitir que os usuários explorem os dados de forma eficiente e identifiquem tendências, padrões e oportunidades de negócios. Além disso, a comunicação eficaz dos insights derivados dos dados é fundamental para orientar a tomada de decisões em toda a organização.

## 2 - Levantamento de requisitos

### 2.1 - Identificação das necessidades e requisitos de dados da Olist Brasil


Dados dos vendedores: 
* A Olist precisa coletar e gerenciar informações detalhadas sobre os vendedores parceiros, como nome, informações de contato, histórico de vendas, políticas de frete, avaliações dos clientes, entre outros. Esses dados são essenciais para gerenciar a relação com os vendedores e avaliar seu desempenho.

Dados de produtos: 
* A Olist requer informações completas e atualizadas sobre os produtos oferecidos pelos vendedores, incluindo categorias, descrições, imagens, preços, variantes, estoque disponível, entre outros. Esses dados são utilizados para listar os produtos nos marketplaces, garantir a precisão das informações apresentadas aos clientes e gerenciar o fluxo de estoque.

Dados de transações: 
* A Olist precisa rastrear e registrar dados sobre as transações realizadas em sua plataforma, como detalhes do pedido, informações do comprador, dados de pagamento, status de entrega, entre outros. Esses dados são fundamentais para fins operacionais, como processamento de pedidos, gerenciamento de logística e análise de desempenho das vendas.

Dados de avaliações e comentários dos clientes: 
* A Olist coleta e analisa as avaliações e os comentários dos clientes sobre os produtos e os vendedores. Esses dados ajudam a monitorar a satisfação do cliente, identificar problemas ou oportunidades de melhoria e fornecer informações valiosas para os vendedores.

Dados de desempenho e análise: 
* A Olist requer dados que permitam acompanhar e analisar o desempenho geral da plataforma, como métricas de vendas, taxas de conversão, número de clientes, índices de satisfação, entre outros. Esses dados são utilizados para avaliar a eficiência operacional, identificar áreas de melhoria e tomar decisões informadas sobre o crescimento e a expansão da empresa.

### 2.3 - Definição dos requisitos funcionais e não funcionais

Em outro momento poderá ser abordada as seguintes etapas:

Requisitos Funcionais:

* Painel de Controle: O sistema deve fornecer um painel de controle intuitivo e amigável que permita aos usuários visualizar métricas de vendas, performance de produtos e estoque disponível, com a capacidade de personalizar os dados exibidos.

* Relatórios Personalizados: O sistema deve oferecer recursos para a geração de relatórios personalizados, permitindo que os usuários extraiam dados específicos de acordo com suas necessidades, como relatórios de desempenho por vendedor ou por categoria de produtos.

* Limpeza e Transformação de Dados: O sistema deve fornecer recursos para limpar e transformar os dados, garantindo a consistência e a integridade dos dados para análises posteriores, com funcionalidades como deduplicação, padronização e validação de dados.

* Segmentação Avançada: O sistema deve permitir a segmentação avançada dos dados com base em critérios personalizados, como histórico de compras, perfil demográfico e comportamento de navegação dos clientes.

Requisitos Não Funcionais:

* Escalabilidade: O sistema deve ser capaz de lidar com um grande volume de dados à medida que a Olist cresce, garantindo um desempenho consistente e tempos de resposta rápidos, independentemente do aumento da carga de dados.

* Segurança e Privacidade: O sistema deve aderir a padrões rigorosos de segurança, garantindo a proteção dos dados confidenciais dos vendedores e clientes. Isso inclui a implementação de criptografia, controle de acesso e conformidade com regulamentações de proteção de dados, como a LGPD.

* Disponibilidade e Confiabilidade: O sistema deve estar disponível e operacional durante a maior parte do tempo, minimizando o tempo de inatividade não planejado. Além disso, deve ser capaz de lidar com possíveis falhas de forma a garantir a continuidade das operações e a recuperação dos dados.

* Integração com Ferramentas de Análise e Visualização: O sistema deve oferecer integração com outras ferramentas populares de análise e visualização de dados, como o Tableau ou o Power BI, permitindo a exploração avançada dos dados e a criação de painéis interativos.

## 3- Arquitetura de dados


### 3.1 - Definição da arquitetura de dados proposta


Camada de Coleta de Dados:

* Integração com os sistemas de vendas da Olist e dos vendedores para coletar dados transacionais, como pedidos, informações de produtos e clientes, e atualizações de estoque.
* Integração com os sistemas de marketing para coletar dados de campanhas, como cliques, conversões e dados de ROI.
* Integração com fontes de dados externas, como dados de mercado e concorrência, para enriquecer a análise.

Camada de Armazenamento de Dados:

* Utilização de um data warehouse para armazenar os dados transacionais e dados agregados necessários para as análises.
* Implementação de um sistema de gerenciamento de banco de dados adequado para garantir a escalabilidade, desempenho e segurança dos dados.

Camada de Processamento e Transformação de Dados:

* Implementação de um pipeline de processamento de dados para limpar, transformar e enriquecer os dados coletados.
* Utilização de técnicas de ETL (Extração, Transformação e Carga) para preparar os dados para análises posteriores.
* Aplicação de lógica de negócios e regras de validação durante o processo de transformação dos dados.

Camada de Análise de Dados:

* Disponibilização de ferramentas de análise e visualização de dados para explorar os dados armazenados e criar relatórios interativos.
* Criação de painéis de controle personalizados que exibem métricas de vendas, desempenho dos produtos, segmentação de clientes e resultados de campanhas de marketing.
* Utilização de algoritmos de análise avançada, como mineração de dados e aprendizado de máquina, para obter insights adicionais dos dados.

Camada de Segurança e Privacidade:

* Implementação de medidas de segurança robustas para proteger os dados confidenciais dos vendedores e clientes, incluindo criptografia de dados em repouso e em trânsito, controle de acesso baseado em funções e monitoramento de atividades suspeitas.
* Adoção de boas práticas de governança de dados e conformidade com regulamentações de proteção de dados, como a LGPD.

### 3.3 - Especificação dos sistemas e tecnologias a serem utilizados (por exemplo, Apache Beam, Google Dataflow, BigQuery)

#### 3.3.1 - Cloud Escolhida

Transient (Transitório):

Raw (Bruto): 
* Nesta fase, os dados brutos são armazenados em seu formato original, sem aplicar transformações significativas. Isso permite preservar os dados em sua forma mais detalhada e possibilita a realização de análises retroativas ou reprocessamento posterior. Os dados brutos são frequentemente armazenados em sistemas de arquivos distribuídos ou em bancos de dados não estruturados.

> Formatos de arquivo não estruturados: Parquet.

> Sistemas de arquivos distribuídos: Google Cloud Storage.

Trusted (Confiança): 
* Nesta fase, os dados são limpos, validados e transformados em um formato confiável. Os dados confiáveis são normalmente armazenados em bancos de dados relacionais ou data warehouses, onde são organizados em esquemas definidos e estruturados. Esses dados estão prontos para análises e consultas de alto desempenho.

> Formatos de arquivo não estruturados: Parquet.

> Sistemas de arquivos distribuídos: Google Cloud Storage.

Refined (Refinado): 
* Nesta fase, os dados são enriquecidos com informações adicionais, agregados, resumidos ou pré-processados para atender a necessidades específicas de análise ou aplicação.                 data1 = pipeline | 'Read Parquet 1' >> parquetio.ReadFromParquet('gs://olist_brasil_project/trusted/customers.parquet')
                data2 = pipeline | 'Read Parquet 2' >> parquetio.ReadFromParquet('gs://olist_brasil_project/raw/customers.parquet')
